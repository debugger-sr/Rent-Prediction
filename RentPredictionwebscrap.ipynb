```python
#Importing libraries and setting up logger for debugging resposes
import requests
import logging
import pandas as pd
import time
from bs4 import BeautifulSoup

from datascience import *
logging.basicConfig()
logging.getLogger().setLevel(logging.DEBUG)
requests_log = logging.getLogger("requests.packages.urllib3")
requests_log.setLevel(logging.DEBUG)
requests_log.propagate = True
```

```python
#understanding website apartments.com
headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.0; WOW64; rv:24.0) Gecko/20100101 Firefox/24.0' }
webpage_response= requests.get('https://www.apartments.com/denton-tx/?bb=ug1x5mvs3Jrly27uN',headers=headers)
print(webpage_response.status_code)
webpage=webpage_response.content
soup = BeautifulSoup(webpage, 'html.parser')
print(soup.prettify())
```

```python
#Function to get data from website
#inputs required:City, Url,Rent,Bed,Bath,Area,Names,Rating
def getdatas(city,i,url):
        a=url.split("?")
        for i in range(1,i):
            if(i==1):
                
                scrapu=url
            else:
                if(city=="denton"):
                    scrapu=a[0]+str(i)+"/?"+a[1]
                else:
                    scrapu=a[0]+str(i)+"/"
            #print(scrapu)
            webpage_response= requests.get(scrapu, headers=headers)
            time.sleep(5)
            
            #print(webpage_response.status_code)
            webpage=webpage_response.content
            soup = BeautifulSoup(webpage, 'html.parser')
            urlss=soup.find_all('article')
            #urlss
            

            for i in urlss:
                    dataurl= i.get('data-url')
                    #print(dataurl)
                    if dataurl != None:
                        #print(dataurl)
                        webpage_response= requests.get(dataurl, headers=headers)
                        time.sleep(5)
                        #print(webpage_response.status_code)
                        webpage=webpage_response.content
                        soup = BeautifulSoup(webpage, 'html.parser')
                        mydivs = soup.find_all("p", {"class": "rentInfoDetail"})
                        #print(mydivs)
                        for n in soup.find_all("h1", {"id": "propertyName"}):
                            names.append(n.getText().strip())
                        #print(names)
                        l=[]
                        for p in soup.find_all("p", {"class": "rentInfoDetail"}):
                            l.append(str(p.getText()))
                        #print(l)
                        if(len(l)==0):
                            rent.append("?")
                            bed.append("?")
                            bath.append("?")
                            area.append("?")
                        else:
                            rent.append(l[0].replace('$', ''))
                            bed.append(l[1].replace('bd', ''))
                            bath.append(l[2].replace('ba', ''))
                            area.append(l[3].replace('sq ft', ''))
                        for k in soup.find_all("div", {"class": "profilePropertyInfoWrapper mortar-wrapper"}):
                            b=list(k.find_all("span", {"class": "reviewRating"}))
                            if(len(b)>0):
                                rating.append( b[0].getText())
                            else:
                                rating.append("?")
                        for n in soup.find_all("section", {"id": "mapSection"}):
                            g=list(n.find_all("div", {"class": "propertyName"}))
                            loc=""
                            for gi in g:
                                for gg in gi:
                                    loc=loc+gg.getText()
                            #print("mapsection",loc.strip())
                            location.append(loc.strip())
                        ls=[]
                        for dg in soup.find_all("h4", {"class": "header-column"}):
                            for do in dg:
                                ls.append(do.getText())
                        #print("ls",ls)
                        if(len(ls)==0):
                            petpolicy.append("?")
                        else:
                            if("Dogs Allowed" in ls or "Cats Allowed" in ls):
                                petpolicy.append(1)
                            else:
                                petpolicy.append(0)
                        emlis=[]
                        for amen in soup.find_all("p", {"class": "amenityLabel"}):
                                #print(amen.getText())
                                emlis.append(amen.getText())
                        #print("emlis",emlis)
                        if(len(emlis)==0):
                            pool.append("?")
                            gated.append("?")
                            fit.append("?")
                            access.append("?")
                            washerdryer.append("?")
                            countertops.append("?")
                            grill.append("?")
                            closet.append("?")
                            continue
                        if "Walk-In Closets" in emlis:
                            closet.append("1")
                        else:
                            closet.append("0")
                            
                        if "Grill" in emlis:
                            grill.append("1")
                        else:
                            grill.append("0")
                            
                        if "Granite Countertops" in emlis:
                            countertops.append("1")
                        else:
                            countertops.append("0")
                            
                        if "In Unit Washer & Dryer" in emlis:
                            washerdryer.append("1")
                        else:
                            washerdryer.append("0")
                            
                        if "Controlled Access" in emlis:
                            access.append("1")
                        else:
                            access.append("0")
                            
                        if "Pool" in emlis:
                            pool.append("1")
                        else:
                            pool.append("0")
                        if "Fitness Center" in emlis:
                            fit.append("1")
                        else:
                            fit.append("0")
                        if "Gated" in emlis:
                            gated.append("1")
                        else:
                            gated.append("0")
                        #print(len(pool),len(fit),len(gated),len(petpolicy))
                            
                        
                                
                        
                                
```
